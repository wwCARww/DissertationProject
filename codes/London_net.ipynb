```python
import statsmodels.api as sm
import seaborn as sns
import statsmodels.formula.api as smf
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

### OLS regression 

```python
# Create a scatter plot of 'car' vs 'smalldata'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=match, x='bigdata', y='smalldata')

# Perform the linear regression
X = match['bigdata'] # independent variable
y = match['smalldata'] # dependent variable

# Add a constant to the independent variable
X = sm.add_constant(X)

# Perform the regression of the dependent on the independent variable
model = sm.OLS(y, X)
results = model.fit()

# Print the regression results
print(results.summary())

# Create a range of values for the fitted line to be plotted over.
x_fit = np.linspace(X['bigdata'].min(), X['bigdata'].max(), 100)
y_fit = results.predict(sm.add_constant(x_fit))

# Add the fitted line to the scatterplot
sns.lineplot(x=x_fit, y=y_fit, color='red')
plt.xlabel('big_data')
plt.ylabel('small_data')
plt.title('Regression of small_data on big_data')

# # Save the figure as a .png image
# plt.savefig('F:/STA/Dissertation/Data/DataValidation_JianiChe/image/Regression of small_data on big_data.png', dpi=150, bbox_inches='tight')  
plt.show()

# Calculate residuals
residuals = results.resid

# Calculate fitted values
fitted = results.fittedvalues

# Define large residuals. Here, let's consider residuals larger than 2 times the standard deviation of residuals as "large"
large_residuals = np.abs(residuals) > 2 * np.std(residuals)

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=fitted, y=residuals, hue=large_residuals, palette={True: 'red', False: 'blue'})
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('Fitted values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()

# Extract and print data points outside the confidence interval: 428/460
outliers = match[large_residuals]
outliers['toid']
```

```python
# Create a scatter plot of 'smalldata' vs 'bigdata'
plt.figure(figsize=(10, 6))

# Perform the linear regression
X = match['bigdata'] # independent variable
y = match['smalldata'] # dependent variable

# Add a constant to the independent variable
X = sm.add_constant(X)

# Perform the regression of the dependent on the independent variable
model = sm.OLS(y, X)
results = model.fit()

# Print the regression results
print(results.summary())

# Create a range of values for the fitted line to be plotted over.
x_fit = np.linspace(X['bigdata'].min(), X['bigdata'].max(), len(match))
y_fit = results.predict(sm.add_constant(match['bigdata']))

# Calculate the confidence interval
predictions = results.get_prediction(sm.add_constant(match['bigdata']))
conf_int = predictions.conf_int(alpha=0.05)

# Scatter plot with points outside the confidence interval colored in red
outside_conf_int = (match['smalldata'] < conf_int[:, 0]) | (match['smalldata'] > conf_int[:, 1])
sns.scatterplot(x=match['bigdata'], y=match['smalldata'], hue=outside_conf_int, 
                palette={True: 'red', False: 'blue'})

# Plotting the regression line
sns.lineplot(x=match['bigdata'], y=y_fit, color='red')

# Plotting the confidence intervals
sns.lineplot(x=match['bigdata'], y=conf_int[:, 0], color='green', linestyle='--')
sns.lineplot(x=match['bigdata'], y=conf_int[:, 1], color='green', linestyle='--')

plt.xlabel('big_data')
plt.ylabel('small_data')
plt.title('Regression of smalldata on bigdata with Confidence Intervals')

# Save the figure
# plt.savefig('F:/STA/Dissertation/Data/DataValidation_JianiChe/image/Regression of small_data on big_data.png', dpi=150, bbox_inches='tight')
plt.show()

# # Extract and print data points outside the confidence interval: 428/460
# outliers = match[outside_conf_int]
# print(outliers) 
```

#### Regression for transformation (log)

```python
# Apply log transformation to 'bigdata' and 'Volume' and add a small constant to avoid log(0)
match['log_bigdata'] = np.log(match['bigdata'] + 1)
match['log_smalldata'] = np.log(match['smalldata'] + 1)

# Drop rows with missing values in 'log_bigdata' or 'log_smalldata'
match_clean = match.dropna(subset=['log_bigdata', 'log_smalldata'])


# Define the dependent variable (y) and the independent variable (X)
X = match_clean['log_bigdata']
y = match_clean['log_smalldata']
# Add a constant to the independent variable
X = sm.add_constant(X)

# Fit the Ordinary Least Squares (OLS) model
model = sm.OLS(y, X)
results = model.fit()

# Print the summary statistics of the regression model
print(results.summary())

# Create a range of values for the fitted line to be plotted over.
x_fit = np.linspace(X['log_bigdata'].min(), X['log_bigdata'].max(), 100)
y_fit = results.predict(sm.add_constant(x_fit))

# Create a scatter plot of 'bigdata' vs 'smalldata'
plt.figure(figsize=(10, 6))
sns.scatterplot(data=match, x='log_bigdata', y='log_smalldata')

# Add the fitted line to the scatterplot
sns.lineplot(x=x_fit, y=y_fit, color='red')
plt.xlabel('log_bigdata')
plt.ylabel('log_smalldata')
plt.title('Regression of log_smalldata on log_bigdata')

# Save the figure as a .png image
plt.savefig('F:/STA/Dissertation/Data/DataValidation_JianiChe/image/Regression of log_smalldata on log_bigdata.png', dpi=150, bbox_inches='tight')  

plt.show()
```

```python
# Convert 'Borough' to a categorical variable
match['Borough'] = match['Borough'].astype('category')

# Get the unique boroughs
boroughs = match['Borough'].unique()

# Create a figure and axes for multiple subplots
fig, axs = plt.subplots(len(boroughs), figsize=(10, 6*len(boroughs)))

for i, borough in enumerate(boroughs):
    # Select only the rows for this borough
    borough_data = match[match['Borough'] == borough]

    # Perform the linear regression
    X = borough_data['bigdata']  # independent variable
    y = borough_data['smalldata']  # dependent variable

    # Add a constant to the independent variable
    X = sm.add_constant(X)

    # Perform the regression of the dependent on the independent variable
    model = sm.OLS(y, X)
    results = model.fit()

    # Print the regression results
    print(f"Regression results for {borough}:\n")
    print(results.summary())
    print("\n---------------------------------------\n")

    # Create a range of values for the fitted line to be plotted over
    x_fit = np.linspace(X['bigdata'].min(), X['bigdata'].max(), 100)
    y_fit = results.predict(sm.add_constant(x_fit))

    # Create a scatter plot of 'bigdata' vs 'smalldata' for this borough
    axs[i].scatter(borough_data['bigdata'], borough_data['smalldata'])
    axs[i].plot(x_fit, y_fit, color='red')  # Add the fitted line
    axs[i].set_title(f"Borough: {borough}")
    axs[i].set_xlabel("bigdata")
    axs[i].set_ylabel("smalldata")

plt.tight_layout()

# Save the figure as a .png image
plt.savefig('F:/STA/Dissertation/Data/DataValidation_JianiChe/image/OLS Regression categorized by Borough.png', dpi=150, bbox_inches='tight')  
plt.show()
```
